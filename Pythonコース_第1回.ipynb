# 4.演算

値と変数を用い、四則演算をはじめ様々な演算を行うことができる。

また、「+」や「-」といった各種演算を行うための記号を演算子と言う。

ここでは、基本的な演算のみ紹介する。

```python
x = 10
y = 4

# 加算
print(x+y)

# 減算
print(x-y)

# 乗算
print(x*y)

# 除算
print(x/y)

# 剰余算
print(x%y)

# べき乗算
print(x**y)

# 切り捨てを行う除算
print(x//y)

# 負数
print(-x)
```

## 4-5. ブール演算

ブール演算自体も組み合わせることが可能。

```python
x = True
y = False
print(x and y) # x かつ y (x, yともにTrueであればTrue)
print(x or y) # x または y(x, yどちらかがTrueであればTrue)
print(not x) # x でない (xがTrueであればFalse、FalseであればTrue)
a = 5
b = 10
c = 15
print((a < b and b < c) or (c < b and b < a))
```

# 5.型変換

演算を行うにあたって、型のルールを理解することは重要である。

ここでは、型の変換について説明する。

## 5-1. 暗黙的型変換

前章の税込金額を求めるプログラムの結果をよく見ると、「.0」まで表示されている。

これはfloat型を意味する表示である。

そこで、変数の型がどのように変化したかを確認する。

```python
# 10%の税込金額を求めるプログラム
price = 100
tax = price * 0.1
print(type(price))
print(type(tax))
price += tax # price = price + tax
print(type(price))
print(price)
```

image.png

先に結論を言うと、int型とfloat型の演算を行うと結果がfloat型に自動で変換される。

そのため、 price * 0.1 を行っている変数taxはfloat型となっており、

int型であった変数priceも price += tax によってfloat型へと変化したので「.0」まで表示された。

このように演算などにより自動的に型が変換されることを**暗黙的型変換**と呼ぶ。

## 5-2. 明示的型変換

int型の値とfloat型の値で演算を行うことで

結果が自動的にfloat型に変換されたが、文字列(str型)は
どうだろうか。

```python
a = 100
b = "50"
c = a + b
print(c)
```

image.png

結果はこのようにエラーとなる。

「+」による演算は数値と文字列両方あるため、 a + b の結果
が 150 という数値を期待しているのか、

 "10050" という文字列を期待しているのか判断できず、Pythonではエラーとなる。

この場合、型を明示的に変換することで解決できる。

```python
a = 100
b = "50"
c = a + int(b) #str型の変数bをint型に変換
print(c)
c = str(a) + b #int型の変数aをstr型に変換
print(c)
```

image.png

このように、 型名(変換元の値や変数) とすることで、型を変換することできる。

なお、下記のように変換先の形式になっていない値はエラーとなる。

```python
x = "ABC"
y = int(x)
print(y)
```

image.png

比較演算についても、型が判定に影響するので適切な変換が必要となる。

```python
a = 100
b = "100"
print(a == b)
print(a == int(b))
```

## 7-3. if〜elif〜else〜

多岐にわたる複雑な分岐を記述する場合、elif文を追加する。

構文は下記の通り。

image.png

先ほどのプログラムに100点の場合のみ「満点合格」と出力するように変更したいとする。

```python
score = int(input("> "))
if score == 100:
  print("満点合格")
elif score >= 60:
  print("合格")
else:
  print("不合格")
```

image.png

さらに負数や100を超える数値が入力された場合に「エラー」と出力させたいとする。

```python
score = int(input("> "))
if score == 100:
  print("満点合格")
elif score < 0 or 100 < score:
  print("エラー")
elif score >= 60:
  print("合格")
else:
  print("不合格")
```

image.png

なお、 101 という値は score >= 60 の条件も満たしているが、

先に記述されているifやelifが優先される。
